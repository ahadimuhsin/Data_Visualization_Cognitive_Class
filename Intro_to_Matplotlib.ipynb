```python
'''Menambah kolom Total di akhir data yang berisi jumlah data dari tahun 1980-2013'''
df_canada['Total']=df_canada.sum(axis=1)
```

```python
'''Histogram'''
#Menggambarkan distribusi frekuensi dari sebuah variabel'

#ingin melihat jumlah imigran tiap negara pada tahun 2013
df_canada['2013'].head()

count, bin_edges = np.histogram(df_canada['2013'])
#np.histogram return 2 values
print (count)
print (bin_edges)
```

```python
'''Plot histogram pada tahun 2013'''
df_canada['2013'].plot(kind='hist', figsize=(8,5), xticks=bin_edges)
plt.title('Histogtam Imigrasi tahun 2013')
plt.ylabel('Jumlah Negara')
plt.xlabel('Jumlah Imigran')

plt.show()
```

```python
'''Distribusi imigrasi dari Denmark, Norwegia, dan Swedia'''
df_canada.loc[['Denmark', 'Norway', 'Sweden'],years]

#membuat histogram
df_canada.loc[['Denmark', 'Norway', 'Sweden'],years].plot.hist()
```

```python
#transpose the data
df_skandinavia = df_canada.loc[['Denmark', 'Norway', 'Sweden'],years].transpose()
df_skandinavia.head()
```

```python
df_skandinavia.plot.hist()
```

```python
df_skandinavia.plot(kind='hist', figsize=(8,5))

plt.title('Histogram Imigrasi Negara-negara Skandinavia')
plt.ylabel('Jumlah Tahun')
plt.xlabel('Jumlah Imigran')

plt.show()
```

```python
#membuat nilai x-tick
count, bin_edges = np.histogram(df_skandinavia, 15)

#visualize histogram
df_skandinavia.plot(kind='hist',
              figsize=(10,6),
              bins=15,
              alpha=0.6,
              xticks=bin_edges,
              color=['red', 'black', 'yellow'])
plt.title('Histogram Imigrasi Negara-negara Skandinavia')
plt.ylabel('Jumlah Tahun')
plt.xlabel('Jumlah Imigran')

plt.show()
```

```python
#untuk mempercantik plot di atas, karena ada yg overlap
xmin = bin_edges[0]-10
xmax = bin_edges[-1]+10

df_skandinavia.plot(kind='hist',
              figsize=(10,6),
              bins=15,
              alpha=0.6,
              xticks=bin_edges,
              color=['red', 'black', 'yellow'],
                   stacked=True,
                   xlim=(xmin,xmax))
plt.title('Histogram Imigrasi Negara-negara Skandinavia')
plt.ylabel('Jumlah Tahun')
plt.xlabel('Jumlah Imigran')

plt.show()
```

```python
##Distribusi untuk negara Yunani, ALbania, dan Bulgaria

df_cof = df_canada.loc[['Greece', 'Albania', 'Bulgaria'], years]

df_cof = df_cof.transpose()

count, bin_edges = np.histogram(df_cof, 15)

#unstacked
df_cof.plot(kind='hist',
           figsize=(10,16),
           bins=15,
           alpha=0.35,
           xticks=bin_edges,
           color=['coral', 'darkslateblue', 'mediumseagreen'],
           stacked=False)
plt.title('Histogram Imigrasi dari Yunani, Albania, dan Bulgaria')
plt.ylabel('Jumlah Tahun')
plt.xlabel('Jumlah Imigran')

plt.show()
```

```python
'''Bar Charts
Di mana panjang dari bar merepresentasikan ukuran dari fitur'''

#Analisis Islandia

df_iceland = df_canada.loc['Iceland', years]
```

```python
df_iceland.plot(kind='bar', figsize=(10,6))

plt.xlabel('Tahun')
plt.ylabel('Jumlah Imigran')
plt.title('Imigran dari Islandia')

plt.show()
```

```python
df_iceland.plot(kind='bar', figsize=(10,6), rot=90)

plt.xlabel('Tahun')
plt.ylabel('Jumlah Imigran')
plt.title('Imigran dari Islandia')

#Anotasi
plt.annotate('',
            xy=(32,70), #letakkan kepala panah di titik ini
            xytext=(28,20), #letakkan inti panah di titik ini
            xycoords='data',
            arrowprops=dict(arrowstyle='->', connectionstyle='arc3', color='blue', lw=2)
            )
plt.annotate('2008 - 2011 Financial Crisis', # text to display
             xy=(28, 30),                    # start the text at at point (year 2008 , pop 30)
             rotation=72.5,                  # based on trial and error to match the arrow
             va='bottom',                    # want the text to be vertically 'bottom' aligned
             ha='left',
            )
plt.show()
```

```python
'''Melihat total imigran ke Kanada dari 15 negara teratas'''
df_canada.sort_values(by='Total', ascending=False, inplace=True)

df_top15= df_canada['Total'].head(15)

df_top15
```

```python
df_top15.plot(kind='barh', figsize=(12,12), color='steelblue')

plt.xlabel('Jumlah Imigran')
plt.title('15 Negara Teratas Imigran di Kanada dari 1980-2013')

#annotasi label nilai untuk tiap negara
for index, value in enumerate(df_top15):
    label = format(int(value),',')
    #letakkan teks pada akhir bar dengan mengurangi 47000 dari x
    #dan 0.1 dari y
    plt.annotate(label, xy=(value-48000, index-0.1),
                color='white')

plt.show()
```
